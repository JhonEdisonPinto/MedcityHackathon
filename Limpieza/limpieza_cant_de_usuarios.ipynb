```python
import pandas as pd

df_cantidad_usuarios = pd.read_csv(r"C:\Users\violeta\Documents\GitHub\MedcityHackathon\Database\cantidad_de_usuarios_por_sitio_dia.csv", encoding="utf-8", sep =",")    
```

```python
# Se borraron las columnas ssid y dia_autenticación
df_cantidad_usuarios.drop(columns=["ssid","dia_autenticación"], inplace=True)
```

```python
df_cantidad_usuarios.head()
```

```python
# Agrupar por mes, año y sede para obtener la cantidad total de registros por mes, año y sede
df_cantidad_usuarios=df_cantidad_usuarios.groupby(["sede","mes","año"])["registros"].agg("sum").reset_index()
```
